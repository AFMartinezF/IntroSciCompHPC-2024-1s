```python
import csv
import matplotlib.pyplot as plt
```

```python
def read_data(file_name):
    with open(file_name, 'r') as file:
        reader = csv.reader(file, delimiter='\t')
        next(reader)
        data = [row for row in reader if len(row)]
    return data
```

```python
seed1 = read_data('randomSeed1.txt')
seed2 = read_data('randomSeed2.txt')
seed5 = read_data('randomSeed5.txt')
pdf = read_data('realPDF.txt')


a0 = [float(row[0]) for row in pdf]
a1 = [float(row[1]) for row in pdf]
xdata = [float(row[0]) for row in seed1]
seed1 = [float(row[1]) for row in seed1]
seed2 = [float(row[1]) for row in seed2]
seed5 = [float(row[1]) for row in seed5]
```

```python
plt.style.use('ggplot')
plt.scatter(xdata, seed1, label='Seed 1',color='g',s=20,alpha=0.5)
plt.scatter(xdata, seed2, label='Seed 2',color='r',s=20,alpha=0.5)
plt.scatter(xdata, seed5, label='Seed 5',color='b',s=20,alpha=0.5)
plt.plot(a0, a1, label='Actual PDF',color='black',linewidth=0.8)
plt.xlabel('Sample values')
plt.ylabel('PDF')
plt.title('PDF for normal distribution')
plt.legend()
plt.savefig('random_pdf.pdf')
plt.show()
```
