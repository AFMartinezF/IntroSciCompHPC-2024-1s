```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos=pd.read_csv('./num.csv',names=['N','S1','S2','Delta'],header=0)
datos.head()
```

```python
plt.style.use('seaborn-v0_8-dark')
plt.grid(True)
plt.title('Error relativo en función de N')
plt.xlabel('N (Cota superior de la sumatoria)')
plt.ylabel('Error relativo △ = |1/S1-S2|')
plt.scatter(datos['N'], datos['Delta'], color='purple', s=0.1, marker='.')
legend  = plt.legend(['Error relativo'], labelcolor='black', loc='upper left', fontsize='large', frameon=True, edgecolor='black', facecolor='white')
legend.legendHandles[0]._sizes = [100]  # Ajusta el tamaño del marcador (en este caso, 50)
legend.legendHandles[0].set_alpha(1)
#plt.savefig('grafica.pdf')  
plt.show()
```
